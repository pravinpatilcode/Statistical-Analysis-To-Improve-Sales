## Task_2

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Read the CSV file
df = pd.read_csv('DS1_C8_S2_SmartCarRiding_Data_Practice.csv')
df
```

```python
# Step 3: Plot the Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['fare_amount'])
plt.title('Box Plot of Fare Amount')
plt.xlabel('Fare Amount')
plt.show()

```

```python
# Step 4: the IQR for outlier treatment
Q1 = df['fare_amount'].quantile(0.25)
Q3 = df['fare_amount'].quantile(0.75)
IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)
```

```python
# Step 5: Filter out outliers
df_filtered = df[(df['fare_amount'] >= lower_bound) & (df['fare_amount'] <= upper_bound)]
df_filtered
```

```python
# Display the filtered DataFrame
df_filtered.head()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Read the CSV file
df = pd.read_csv('DS1_C8_S2_SmartCarRiding_Data_Practice.csv')
df
```

```python
# Step 3: Define the sample sizes
sample_sizes = [10, 15, 30, 200, 1000, 5000, 10000, 50000, 100000]
sample_sizes
```

```python
# Step 4: Loop through sample sizes and plot distributions
for size in sample_sizes:
    # Check if the sample size is larger than the dataset
    if size > len(df):
        print(f"Sample size {size} is larger than the dataset size.")
        continue
```

```python
# Generate a random sample of the specified size
sample = df['fare_amount'].sample(n=size, random_state=1)
sample
```

```python
# Plot the distribution of the sample
plt.figure(figsize=(12, 6))
sns.histplot(sample, kde=True, bins=30, color='skyblue')
plt.title(f'Distribution Plot for Sample Size = {size}')
plt.xlabel('Fare Amount')
plt.ylabel('Frequency')
plt.show()
```

## Interpretation:-Samples with sizes greater than or equal to 200 typically exhibit distributions that more closely approximate a normal distribution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DS1_C8_S2_SmartCarRiding_Data_Practice.csv')
data
```

```python
## Calculate Central Tendency and Dispersion for Entire Data

entire_data_mean = data.mean()
entire_data_median = data.median()
entire_data_std = data.std()
entire_data_variance = data.var()
```

```python
print("Entire Data - Mean:")
print(entire_data_mean)
print("Entire Data - Median:")
print(entire_data_median)
print("Entire Data - Standard Deviation:")
print(entire_data_std)
print("Entire Data - Variance:")
print(entire_data_variance)
```

```python
sample_data = data.sample(200, random_state=1)
sample_data
```

```python
## Calculate Central Tendency and Dispersion for Sampled Data
sample_mean = sample_data.mean()
sample_median = sample_data.median()
sample_std = sample_data.std()
sample_variance = sample_data.var()
```

```python
print("Sample Data - Mean:")
print(sample_mean)
print("Sample Data - Median:")
print(sample_median)
print("Sample Data - Standard Deviation:")
print(sample_std)
print("Sample Data - Variance:")
print(sample_variance)
```

```python
# Check for Central Limit Theorem (CLT)
# A visual inspection of the sample distribution can be helpful

plt.hist(sample_data.mean(), bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Sample Mean')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```


## Justify Based on Results :-  If the sample mean is close to the population mean and the sample distribution of the means is approximately normal, this supports the Central Limit Theorem and confirms that the sample data is suitable for analysis.
