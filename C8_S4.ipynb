# Task 1
a.What is the null and alternate hypothesis to determine if the mean score for the class test would be above 71.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
H0:μ≤71
Alternate Hypothesis (H1): The mean score of the students is greater than 71.
𝐻1:𝜇>71
H1:μ>71
    # Right tailed test
```

```python
# b.	For alpha=0.1, what is the critical value of the hypothesis test?
```

```python
alpha = 0.1
n = 6  # sample size
critical_value = stats.t.ppf(1 - alpha, df=n-1)
critical_value
```

```python
# c.	What are the test statistic and p-value for the hypothesis test?
```

```python
scores = np.array([61, 93, 75, 67, 84, 95])

sample_mean = np.mean(scores) # Calculate sample mean and standard deviation
sample_std = np.std(scores, ddof=1)  # sample standard deviation
n = len(scores)

t_statistic = (sample_mean - 71) / (sample_std / np.sqrt(n)) # Calculate t-statistic

p_value = 1 - stats.t.cdf(t_statistic, df=n-1) # Calculate p-value for the right-tailed test

sample_mean, sample_std, t_statistic, p_value
```

```python
# d.	Draw the conclusion for the hypothesis test and plot the distribution.
```

```python

if t_statistic > critical_value:
    conclusion_critical = "Reject the null hypothesis (mean score is above 71)."
else:
    conclusion_critical = "Fail to reject the null hypothesis (no evidence mean score is above 71)."

# Conclusion based on p-value
if p_value < alpha:
    conclusion_p_value = "Reject the null hypothesis (mean score is above 71)."
else:
    conclusion_p_value = "Fail to reject the null hypothesis (no evidence mean score is above 71)."

# Print conclusions
conclusion_critical, conclusion_p_value

# Plotting the t-distribution
x = np.linspace(40, 100, 300)
y = stats.t.pdf(x, df=n-1)

plt.plot(x, y, label='t-distribution', color='blue')
plt.fill_between(x, y, where=(x >= critical_value + 71), color='red', alpha=0.5, label='Rejection Region')

plt.axvline(x=71, color='green', linestyle='--', label='Mean under H0 (71)')
plt.axvline(x=sample_mean, color='orange', linestyle='--', label='Sample Mean')

plt.title('Hypothesis Test for Mean Score')
plt.xlabel('Score')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```
